#### 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
num = int(input("Enter Any Number Here : "))
if num > 0 :
    print("Number Is Positive")
    if num % 2 == 0 :
        print("Number Is also Even")
    else :
        print("Number Is Odd")
elif num == 0 :
    print("Number Is Zero")
else :
    print("Numer Is Negative")
```

### Section C: Exploring Pandas

#### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import numpy as np
import pandas as pd

data = {
    "Name" : ["Rio", "Rio", "Neha", "Neha", "Karan", "Smit", "Arun", "Kuldeep", "Maulik", "Khushi"], 
    "Age" : [21, 21, 24, 24, 21, 22, 20, 21, 21, 22],
    "City" :["Mumbai", "Mumbai", "Pune", "Pune", "Surat", "Delhi", "Mumbai", "Jaipur", "Indore", "Ahmedabad"],
     "Salary": [50000, 50000, 55000, 55000, 65000, 50000, np.nan, 70000, np.nan, 55000]
}

df = pd.DataFrame(data) #Creating Dataframe
df
```

```python
df.info() #info
```

```python
df.describe() #Describe
```

```python
df[["Name", "City"]] #Select 'Name' and 'City' columns
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())#Fill any missing values in 'Salary' column with the mean
df
```

```python
new_df = df.drop_duplicates() #Remove any duplicate rows
new_df
```
